```python
import json
import pandas as pd
import numpy as np
```

```python
# Read the Overall Rankings at 2023
with open("team_rank.json", "r") as file:
    team_rank = json.load(file)

# Read the Power Rankings for each tournament
with open("tournament_rank.json", "r") as file:
    tournament_rank = json.load(file)
```

# Tournament Rankings
The tournament name is the key for the tournament rank dictionary.  Use it to access whichever tournament you wish to check.  MSI 2023 has been loaded as an example

```python
# Extract the specific tournament
tournament = tournament_rank['msi_2023']

# Convert the dictionary to a DataFrame
df_tournament = pd.DataFrame(list(tournament.items()), columns=['Team', 'Ranking'])

# Sort the DataFrame by 'Ranking' in ascending order
df_sorted = df_tournament.sort_values(by='Ranking').reset_index(drop=True)

df_sorted
```

# End of 2023 Rankings
Below we get the rankings for the end of 2023.  I have populated a list with the team names of the teams participating in worlds.  You can change the team names and run the cell 2 cells below to get the results.

```python
# Checking worlds 2023 teams - can substitute with any slugs of teams to search the whole ranking
list_of_teams = ['jd-gaming','lng-esports','bilibili-gaming','geng','kt-rolster','t1','dwg-kia','weibo-gaming',
                 'fnatic','g2-esports','mad-lions', 'team-bds', 'cloud9', 'nrg-esports', 'team-liquid', 'gam-esports']
```

```python
# Extract rankings for the teams in team_list
specific_team_ranks = {team: team_rank[team] for team in list_of_teams if team in team_rank}

# Sort the teams based on their rankings
sorted_teams = sorted(specific_team_ranks.items(), key=lambda x: x[1])

# Convert to DataFrame
rank_df = pd.DataFrame(sorted_teams, columns=['Team', 'Rank'])

# Display the DataFrame
rank_df
```

```python
# End of 2023 Power Rankings
rank_df = pd.DataFrame(list(team_rank.items()), columns=['Team', 'Rank']).sort_values(by='Rank')
rank_df.style.set_caption("Team Rankings")
```
